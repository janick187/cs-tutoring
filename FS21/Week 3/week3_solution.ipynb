## Distance Caluclator

Subsequently, we are going to implement our program that asks the user to enter two directions and then calculates the distance between these two directions using the Google Maps API. Finally, the program presents the result to the user.

To access the Google Maps API we need an API Key (so that google is sure we are not a random robot or so). Please feel free to [generate and use your own personal API key](https://developers.google.com/maps/documentation/embed/get-api-key), maybe you can re-use it one time in the future ;-)

### Calculate distance
Let's use the Google Maps API to calcualte the distance between the two directions.

```
# your code here
import requests

def calculateDistance(start, destination):
  base_url = "https://maps.googleapis.com/maps/api/distancematrix/json?"
  api_key = ""

  if api_key == "":
    print("ERROR: You must specifcy an API key!")
    return

  # Example Request URL https://maps.googleapis.com/maps/api/distancematrix/json?origins=Vancouver+BC|Seattle&destinations=San+Francisco|Victoria+BC&mode=bicycling&language=fr-FR&key=AIzaSyCMTkl9rQgCfwHwZfiDlYK7HS7gRmJZtZ8
  url = "{}origins={}&destinations={}&key={}".format(base_url,start,destination, api_key)

  response_dict = requests.get(url).json()

  origin = response_dict["origin_addresses"][0]
  destination = response_dict["destination_addresses"][0]
  distance = response_dict["rows"][0]["elements"][0]['distance']['text']

  print("The distance between {} and {} is {}".format(origin, destination, distance))
```

### User Input
Let's get the information from the user we need. This is a source and destination direction.

```
# your code here
start = input("Where do you leave from? (enter number, street and place with a + in between) ")
destination = input("Where do you go? (enter number, street and place with a + in between) ")

calculateDistance(start, destination)
```
